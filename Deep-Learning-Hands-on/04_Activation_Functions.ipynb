## ***Activation Functions***
***

In this notebook, I tackled a few of the activation functions and discuss their roles. We use different activation functions for different cases, and understanding how they work can help you properly pick which of them is best for your task. The activation function is applied to the output of a neuron (or layer of neurons), which modifies outputs. We use activation functions because if the activation function itself is nonlinear, it allows for neural networks with usually two or more hidden layers to map nonlinear functions.

In general, your neural network will have two types of activation functions. The first will be the activation function used in hidden layers, and the second will be used in the output layer. Usually, the activation function used for hidden neurons will be the same for all of them, but it does not have to.

***
#### ***The Step Activation Function***
***

the purpose Step Activation Function serves is to mimic a neuron **“firing”** or **“not firing”**based on input information. The simplest version of this is a step function. In a single neuron, if the **weights * inputs + bias**​ results in a value greater than 0, the neuron will fire and output a 1 otherwise, it will output a 0. This activation function has been used historically in hidden layers, but nowadays, it is rarely a choice. 


***
#### ***The Linear Activation Function***
***

A **linear function** is simply the **equation of a line**. It will appear as a straight line when graphed, where **y=x** and the output value equals the input.

This activation function is usually applied to the last layer’s output in the case of a regression model a model that outputs a scalar value instead of a classification.

#### **The Sigmoid Activation Function***
***

The problem with the step function is **it’s not very informative**. When we get to training and network optimizers, you will see that the way an optimizer works is by assessing individual impacts that weights and biases have on a network’s output. The problem with a step function is that it’s less clear to the optimizer what these impacts are because there’s very little information gathered from this function. **It’s either on (1) or off (0)**. It’s hard to tell how **“close”** this step function was to activating or deactivating. Maybe it was very close, or maybe it was very far. In terms of the final output value from the network, it does not matter if it was close​ to outputting something else. Thus, when it comes time to optimize weights and biases, it’s easier for the optimizer if we have activation functions that are more granular and informative. 

The original, more granular, activation function used for neural networks was the Sigmoid activation function, which looks like: 

```python

from IPython.display import Image
Image(filename='sfg.png') 
```

This function returns a value in the range of 0 for negative infinity, through 0.5 for the input of 0, and to 1 for positive infinity.

As mentioned earlier, with **“dead neurons,”** it’s usually better to have a more granular approach for the hidden neuron activation functions. In this case, we are getting a value that can be reversed to its original value, the returned value contains all the information from the input, contrary to a function like the step function, where an input of 3 will output the same value as an input of 300,000. The output from the Sigmoid function, being in the range of 0 to 1, also works better with neural networks especially compared to the range of the negative to the positive infinity and adds nonlinearity. The importance of nonlinearity will become more clear shortly in this chapter. The Sigmoid function, historically used in hidden layers, was eventually replaced by the **Rectified Linear Units activation function (or ReLU)**.

***

### ***Why Use Activation Functions?***
***

Activation functions are a vital component of neural networks because they introduce non-linearity, enabling the network to learn and model complex patterns in data. Without activation functions, even a deep neural network would behave like a linear model, severely limiting its ability to solve real-world, non-linear problems. By introducing non-linearity, activation functions allow the network to capture intricate relationships in data and support hierarchical feature learning. This enables deeper layers to build upon the features learned by earlier layers, leading to better performance in tasks such as image recognition, natural language processing, and speech recognition.

Activation functions also help control the output range of neurons, which can stabilize learning and prevent issues like exploding or vanishing gradients. They are essential for gradient-based optimization, as smooth and differentiable activation functions facilitate effective weight updates during backpropagation. Different activation functions serve specific purposes: for example, sigmoid functions are ideal for binary classification, softmax is used in multi-class classification, and ReLU is commonly used in hidden layers for efficiency and to avoid saturation problems.

In addition to improving the network's capacity to model complex data, activation functions also contribute to better generalization and reduced overfitting. Variants like leaky ReLU help address issues like dead neurons, further enhancing the model’s performance. By carefully selecting the right activation function for a given task and architecture, neural networks can unlock their full potential to solve real-world problems effectively.

***
### ***Linear Activation in the Hidden Layers***
***

Using a linear activation function in the hidden layers of a neural network significantly limits the network's capabilities. Linear activation means the output of each neuron is a direct linear transformation of its input. While this approach simplifies computations, it undermines the core advantage of neural networks: their ability to model complex, non-linear relationships in data.

The primary issue is that a network with only linear activation functions, regardless of the number of layers, behaves like a single-layer linear model. This is because a composition of linear transformations remains linear. As a result, the network cannot capture non-linear patterns, making it ineffective for solving complex tasks such as image recognition, speech processing, or natural language understanding.

Moreover, without non-linearity, the network loses the ability to learn hierarchical features. Neural networks are designed to build increasingly abstract representations of data as information passes through hidden layers. Linear activations prevent this process, as each layer essentially performs the same operation, leading to redundancy and no added value from deeper architectures.

Although linear activation is unsuitable for hidden layers, it is occasionally used in the output layer for specific tasks. For instance, in regression problems where the goal is to predict a continuous value, a linear output activation is appropriate. In such cases, the network benefits from linear activation to produce unrestricted output values directly. However, for hidden layers, introducing non-linearity through functions like ReLU, sigmoid, or tanh is essential to unlock the true potential of neural networks.

***
### ***ReLU Activation in a Pair of Neurons***
***

The Rectified Linear Unit (ReLU) activation function, defined as **𝑓(𝑥)=max⁡(0,𝑥)**, is widely used in neural networks due to its simplicity and effectiveness. When applied to a pair of neurons, ReLU enables the network to capture non-linear patterns while maintaining computational efficiency. Each neuron processes an input through a linear transformation followed by the ReLU function. For example, if a neuron computes 𝑧=𝑊⋅𝑥+𝑏, the ReLU activation outputs **𝑎=max⁡(0,𝑧)**, effectively setting all negative values to zero and passing positive values unchanged.

In a pair of neurons, this process allows each neuron to activate independently, enabling diverse and complementary feature extraction. For instance, one neuron might output a positive value while the other outputs zero, demonstrating the sparsity introduced by ReLU. Sparsity not only improves computational efficiency but also helps reduce overfitting by ensuring that only a subset of neurons are active at any given time. Additionally, ReLU facilitates efficient gradient flow during backpropagation, avoiding the vanishing gradient problem common in sigmoid or tanh activations.

Consider a scenario where two neurons process the same input. If the first neuron outputs a positive value, such as 14, and the second outputs a negative value, such as -10, the ReLU activation results in outputs of 14 and 0, respectively. This non-linearity allows the network to better model complex relationships in the data, even with simple architectures.

By using ReLU activation in a pair of neurons, neural networks benefit from improved learning capabilities, efficient computation, and the ability to extract meaningful patterns, making it a fundamental choice in modern deep learning architectures.

***
### ***ReLU Activation in the Hidden Layers***
***

The Rectified Linear Unit (ReLU) activation function, defined as **f(x)=max(0,x)**, is a fundamental component of modern neural networks and is widely used in **hidden layers**. **ReLU introduces non-linearity to the network**, enabling it to model complex relationships in data. Without a non-linear activation like ReLU, even a deep neural network would behave like a linear model, limiting its ability to solve real-world problems that involve intricate patterns.

ReLU is particularly effective in hidden layers due to its simplicity and computational efficiency. It performs a straightforward operation: outputting the input if it is positive and returning zero if it is negative. This simplicity allows ReLU to be computed quickly, making it suitable for large-scale networks with many hidden layers. Additionally, ReLU helps address the vanishing gradient problem that occurs in activation functions like sigmoid or tanh. For positive inputs, the gradient of ReLU is 1, which ensures stable and efficient weight updates during backpropagation.

Another advantage of ReLU in hidden layers is its ability to create sparse activations. By setting negative inputs to zero, ReLU ensures that only a subset of neurons are active at any given time. This sparsity not only reduces computational complexity but also helps prevent overfitting by promoting simpler representations.

Overall, the use of ReLU in hidden layers is key to the success of deep learning models. It enables the network to capture non-linear patterns, supports efficient training, and contributes to better generalization, making it a cornerstone of modern neural network architectures.

### ***The Softmax Activation Function***
***

The Softmax activation function is commonly used in the output layer of neural networks, particularly for multi-class classification problems. It converts raw output scores (also called logits) from the network into probabilities that sum to 1, which makes it easier to interpret the network’s predictions. The Softmax function is mathematically defined as:

$$
\sigma(z_i) = \frac{e^{z_i}}{\sum_{j=1}^N e^{z_j}}
$$

Where:
- \(z_i\) is the raw score (logit) for the \(i\)-th class.
- \(N\) is the total number of classes.
- \(e^{z_i}\) ensures all outputs are positive.
- The denominator normalizes the values so that the probabilities sum to 1.

By applying the Softmax function, the network emphasizes the largest logit, which corresponds to the predicted class, while suppressing the smaller logits, thus creating a clear distinction among the predicted probabilities. For example, given logits **[1.2,0.9,0.3]**, the Softmax function would output the probabilities associated with each class, with the largest logit receiving the highest probability. This makes Softmax ideal for problems where you need to classify inputs into one of several categories.
