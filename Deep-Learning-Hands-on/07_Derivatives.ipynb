# ***Derivatives***
***

*A **derivative** measures how a function changes as its input changes. In the context of neural networks:*

- The loss function is our function of interest.

- The weights and biases are the inputs to this function that we want to adjust to minimize the loss.

**Here's a basic example:**

Imagine a simple function ùëì(ùë•)=ùë•2. The derivative of this function, denoted as ùëì‚Ä≤(ùë•), tells us how ùëì(ùë•) changes with respect to ùë•.

For ùëì(ùë•) = ùë•2, the derivative ùëì‚Ä≤(ùë•) = 2ùë•. This means that the rate of change of ùëì(ùë•) with respect to ùë• is 2ùë•.

**In neural networks, we deal with more complex functions, but the principle remains the same:**

**Partial derivatives:** When dealing with functions of multiple variables (e.g., weights and biases), we use partial derivatives. They measure how the function changes with respect to one variable, holding the others constant.

**Gradient:** The gradient is a vector of partial derivatives, representing the direction and rate of the fastest increase of the loss function. In gradient descent, we move in the opposite direction of the gradient to minimize the loss.

For example, if our loss function ùêø depends on weights ùë§ and biases ùëè, we calculate the partial derivatives ‚àÇùêø‚àÇùë§ and ‚àÇùêø‚àÇùëè. These values tell us how to adjust ùë§ and ùëè to reduce the loss.

### ***The Impact of a Parameter on the Output***
***

To understand **partial derivatives**, we need to start with derivatives, which are a special case of partial derivatives they are calculated from functions taking single parameters.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return 2 * x

# Create an array of input values from 0 to 4
x = np.array(range(5))

# Calculate the function output for each input value
y = f(x)

# Print the input and output values
print(x)
print(y)

# Plot the result
plt.plot(x, y)
plt.xlabel('Input (x)')
plt.ylabel('Output (y)')
plt.title('Impact of x on y for the function y = 2x')

plt.show()
```

### ***The Slope***
***

This looks like an output of the f(x) = 2x function, which is a line. How might you define the impact that x will have on y? Some will say, **‚Äúy is double x‚Äù** Another way to describe the impact of a linear function such as this comes from algebra: the slope. **‚ÄúRise over run‚Äù** might be a phrase you recall from school. 
The slope of the line ùë¶=2ùë• can be found using the formula for the slope:
$$
slope =Œîùë¶/Œîùë•
$$
Where:

- Œîùë¶ is the change in ùë¶

- Œîùë• is the change in ùë•

It is change in y divided by change in x, or, in math  delta y divided by delta x. What‚Äôs the slopeof f(x) = 2x then? To calculate the slope, first we have to take any two points lying on the function‚Äôs graph andsubtract them to calculate the change. Subtracting the points means to subtract their x and y dimensions respectively. Division of the change in y by the change in x returns the slope:

- Point 1: (0, 0)

- Point 2: (1, 2)

$$
Œîùë• = p_{2x} - p_{1x} = 1 - 0 = 1
$$
$$
Œîy = p_{2y} - p_{1y} = 2 - 0 = 2
$$
$$
slope =Œîùë¶/Œîùë• = 2/1 = 1
$$


Continuing the code, we keep all values of x in a single-dimensional NumPy array, x, and all results in a single-dimensional array, y. To perform the same operation, we‚Äôll take x[0] and y[0] for the first point, then x[1] and y[1] for the second one. Now we can calculate the slope between them:

```python

print(( y[1] - y[0] ) / ( x[1] - x[0] ))
```

It is not surprising that the slope of this line is 2. We could say the measure of the impact that x has on y is 2. We can calculate the slope in the same way for any linear function, including linear functions that aren‚Äôt as obvious. What about a nonlinear function like 
$$
f(x)=2_{x^2}
$$

**How might we measure the impact that x has on y in this nonlinear function?** 

Calculus proposes that we measure the slope of the **tangent line** at x (for a specific input value to the function), giving us the **instantaneous slope** (slope at this point), which is the **derivative**. The **tangent line** is created by drawing a line between two points that are ‚Äúinfinitely close‚Äù on a curve, but this curve has to be differentiable at the derivation point. This means that it has to be continuous and smooth (we cannot calculate the slope at something that we could describe as a **‚Äúsharp corner‚Äù**, since it contains an infinite number of slopes). Then, because this is a curve, there is no single slope. Slope depends on where we measure it. To give an immediate example, we can approximate a derivative of the function at x by using this point and another one also taken at x, but with a very small delta added to it, such as 0.0001. This number is a common choice as it does not introduce too large an error (when estimating the derivative) or cause the whole expression to be numerically unstable (Œîx might round to 0 due to floating-point number resolution). This lets us perform the same calculation for the slope as before, but on two points that are very close to each other, resulting in a good approximation of a slope at x.

```python

p2_delta = 0.0001

x1 = 1

x2 = x1 + p2_delta # add delta

y1 = f(x1) # result at the derivation point

y2 = f(x2) # result at the other, close point

approximate_derivative = (y2-y1)/(x2-x1)

print(approximate_derivative)
```

As we will soon learn, the derivative of **2x^2** at **x=1** should be exactly 4. The difference we see (~4.0002) comes from the method used to compute the tangent. We chose a delta small enough to approximate the derivative as accurately as possible but large enough to prevent a rounding error. To elaborate, an infinitely small delta value will approximate an accurate derivative; however, the delta value needs to be numerically stable, meaning, our delta can not surpass the limitations of Python‚Äôs floating-point precision (can‚Äôt be too small as it might be rounded to 0 and, as we know, dividing by 0 is **‚Äúillegal‚Äù**). Our solution is, therefore, restricted between estimating the derivative and remaining numerically stable, thus introducing this small but visible error.

### ***The Numerical Derivative***
***

Numerical differentiation is a powerful tool when we need to approximate the derivative of a function based on discrete data points or when an analytical expression of the derivative is not available.

To estimate the derivative of a function ùëì(ùë•) at a point ùë•, we use the following formula:

$$ f‚Ä≤(x) = \frac{ùëì(ùë•+‚Ñé)‚àíùëì(ùë•)}{h} $$

where:

- ‚Ñé
 is a small number representing the difference between two close points.

- ùëì(ùë•+‚Ñé)‚àíùëì(ùë•) is the change in ùë¶ values.

- ‚Ñé is the change in ùë• values.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 2 * x ** 2

# Define a small h value
h = 1e-5

# Calculate the numerical derivative at x = 2
x = 2
numerical_derivative = (f(x + h) - f(x)) / h

# Print the numerical derivative
print(f"The numerical derivative at x = {x} is approximately {numerical_derivative}")

# Visualization
x_values = np.linspace(0, 4, 500)
y_values = f(x_values)
plt.plot(x_values, y_values, label='y = 2x')

# Tangent line at x = 2
tangent_x = np.array([x, x + h])
tangent_y = np.array([f(x), f(x + h)])
plt.plot(tangent_x, tangent_y, 'ro-', label='Tangent line at x = 2')

plt.xlabel('Input (x)')
plt.ylabel('Output (y)')
plt.title('Numerical Derivative of y = 2x at x = 2')
plt.legend()
plt.grid(True)
plt.show()
```

For this simple function, f(x) = 2x2, we did not pay a high penalty by approximating the derivative(i.e., the slope of the tangent line) like this, and received a value that was close enough for our needs. 

The problem is that the actual function employed in our neural network is not so simple. The loss function contains all of the layers, weights, and biases it is an absolutely massive function operating in multiple dimensions! Calculating derivatives using **numerical differentiation** requires multiple forward passes for a single parameter update (we will talk about parameter updates in chapter 10). We need to perform the forward pass as a reference, then update a single parameter by the delta value and perform the forward pass through our model again to see the change of the loss value. Next, we need to calculate the derivative and revert the parameter change that we made for this calculation. We have to repeat this for every weight and bias and for every sample, which will be very time-consuming. We can also think of this method as brute-forcing the derivative calculations. To reiterate, as we quickly covered many terms, the derivative is the slope of the tangent line for a function that takes a single parameter as an input. We will use this ability to calculate the slopes of the loss function at each of the weight and bias points this brings us to the multivariate function, which is a function that takes multiple parameters and is a topic for the the partial derivative.

### ***The Analytical Derivative***
***

Now that we have a better idea of what a **derivative** is, how to calculate the numerical (also called **universal**) derivative, and why it‚Äôs not a good approach for us, we can move on to the Analytical Derivative, the actual solution to the derivative that we will implement in our code.
    
In mathematics, there are two general ways to solve problems: **numerical and analytical methods**. Numerical solution methods involve coming up with a number to find a solution, like the above approach with approximate_derivative. The numerical solution is also an approximation. On the other hand, the analytical method offers the exact and much quicker, in terms of calculation, solution. However, identifying the analytical solution for the derivative of a given function, as we‚Äôll quickly learn, will vary in complexity, whereas the numerical approachnever gets more complicated it‚Äôs always calling the method twice with two inputs to calculate the approximate derivative at a point. Some analytical solutions are quite obvious, some can be calculated with simple rules, and some complex functions can be broken down into simpler parts and calculated using the so-called **chain rule**. We can leverage already-proven derivative solutions for certain functions, and others like our loss function can be solved with combinations of the above.
    
To compute the derivative of functions using the analytical method, we can split them into simple, elemental functions, finding the derivatives of those and then applying the **chain rule**, which we will explain soon, to get the full derivative. To start building an intuition, let‚Äôs start with simple functions and their respective derivatives.

The derivative of a simple constant function:
$$
f(x) = 1  = \frac{d}{dx} f(x) = \frac{d}{dx} 1 = 0
$$
